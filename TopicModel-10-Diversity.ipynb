```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from gensim import corpora, models

# Load data
X  = np.load('data/one_hot_coco.npy')
    
lda = models.ldamodel.LdaModel.load("data/hier104-10-topics.model")
```

```python
# Load results
with open("data/hier104-10-topic-res.txt", "rb") as fp:
    img_by_topic = pickle.load(fp)
```

# 10 Topic Model
### Select 400 images from each topic and visualize their "feature vector"


```python
topic_mat = lda.get_topics()

fig, ax = plt.subplots(nrows=10, ncols=1, figsize=(18,18))

i = 0
for row in ax:
    ind = [x[0] for x in img_by_topic[i]]
    row.imshow(np.multiply(X[ind[:400],:], topic_mat[i]))
    i += 1
```

```python
# Load results
with open("data/hier104-0-4-topic-res.txt", "rb") as fp:
    sub_img_by_topic = pickle.load(fp)

sub_lda = models.ldamodel.LdaModel.load("data/hier104-0-4-topics.model")
sub_topic_mat = sub_lda.get_topics()

fig, ax = plt.subplots(nrows=4, ncols=1, figsize=(18,18))

i = 0
for row in ax:
    sub_ind = [x[0] for x in sub_img_by_topic[i]]
    row.imshow(np.multiply(X[sub_ind[:1000],:], sub_topic_mat[i]))
    i += 1
```
